```python
import os
import xml.etree.ElementTree as ET

# Đường dẫn tới thư mục chứa hình ảnh và file XML
image_folder = 'D:/School/Traffic data/FrameOfRedLights'
xml_folder = 'D:/School/Traffic data/RedLightRectBoxed'

# Hàm kiểm tra xem ảnh đã được gán nhãn RectBox hay chưa
def is_rectbox_present(xml_file):
    if not os.path.exists(xml_file):
        return False
    
    tree = ET.parse(xml_file)
    root = tree.getroot()
    for obj in root.findall('object'):
        if obj.find('name').text == 'RedLight':
            bndbox = obj.find('bndbox')
            if bndbox is not None:
                return True
    return False

# Lặp qua tất cả các file hình ảnh và kiểm tra RectBox
for filename in os.listdir(image_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Chỉ xử lý file ảnh
        xml_filename = os.path.join(xml_folder, os.path.splitext(filename)[0] + '.xml')
        if is_rectbox_present(xml_filename):
            print(f"{filename} has been RectBox.")
        else:
            print(f"{filename} hasn't RectBox.")
            # Tạo file XML mới nếu ảnh chưa được gán nhãn
            create_xml(filename, xml_filename)

print("Done!")
```
